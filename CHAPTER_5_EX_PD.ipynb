```
def exercise_ten(s,t,c0):

    # state variables
    c1 = s[0]
    c2 = s[1]
    c1sp = s[2]
    c2sp = s[3]

    # physical model 
    dc1dt = -((F0s + FRs)/V1s + k1)*c1 + FRs/V1s*c2 + F0s/V1s*c0

    dc2dt = -((F0s + FRs)/V2s + k2)*c2 + (F0s + FRs)/V2s*c1

    # state-space model 
    dc1spdt = -((F0s + FRs)/V1s + k1)*(c1sp - c1s) + FRs/V1s*(c2sp - c2s) + F0s/V1s*(c0 - c0s)

    dc2spdt = -((F0s + FRs)/V2s + k2)*(c2sp - c2s) + (F0s + FRs)/V2s*(c1sp - c1s)

    return [dc1dt, dc2dt, dc1spdt, dc2spdt]
```

```
def heater(s,t,u):

    # state variables 
    T = s[0]
    Tj = s[1]
    Tsp = s[2]
    Tjsp = s[3]

    # inputs 
    Fj = u[0]
    F = u[1]
    Ti = u[2]
    Tjin = u[3]

    # physical model

    # energy balance in the tank 
    dTdt = F/Vs*(Ti - T) + UA/(rhocp*Vs)*(Tj - T)

    # energy balance in the jacket 
    dTjdt = Fj/Vjs*(Tjin - Tj) - UA/(rhocp*Vjs)*(Tj - T)

    # state space model

    # energy balance in the tank 
    dTspdt = -(Fs/Vs + UA/(rhocp*Vs))*(Tsp - Ts) + UA/(rhocp*Vs)*(Tjsp - Tjs) + (Tis - Ts)/Vs*(F - Fs) + Fs/Vs*(Ti - Tis)

    # energy balance in the jacket 
    dTjspdt = UA/(rhocp*Vjs)*(Tsp - Ts) - (Fjs/Vjs + UA/(rhocp*Vjs))*(Tjsp - Tjs) + (Tjins- Tjs)/Vjs*(Fj - Fjs) + Fjs/Vjs*(Tjin - Tjins)

    return [dTdt, dTjdt, dTspdt, dTjspdt]
```

```
# nominal values 
Fs = 1 # ft3/min
Fjs = 1.5 # ft3/min
Tis = 50 # °F
Ts = 125 # °F
Vs = 10 # ft3
Tjins = 200 # °F
Tjs = 150 # °F
Vjs = 1 # ft3 
```

```
# parameters 
UA = 183.9 # Btu/(°F min)
rhocp = 61.3 # Btu/(°F ft3)
```

```
# initial conditions 
s = [Ts, Tjs, Ts, Tjs]

# inputs 
u = [Fjs, Fs,Tis, Tjins]
```

```
# integrqtion time 
t = np.linspace(0,10,100)
```

```
out = odeint(heater, s, t, args=(u,))
```

```
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
plt.plot(t, out[:,0], t, out[:,1])
plt.xlabel("time (min)")
plt.legend(["T","Tj"])
plt.subplot(2,1,2)
plt.plot(t, out[:,2], t, out[:,3])
plt.xlabel("time (min)")
plt.legend(["Tsp","Tjsp"])
```

```
# lets to change the input Fj from 1.5 to 1.75 at minute 5
# input vectors 
Fj = np.ones(len(t))*Fjs
#Fj[50:100] = 1.75
F = np.ones(len(t))*Fs
#F[25:75] = 1.5
Ti = np.ones(len(t))*Tis
#Ti[40:80] = 75
Tjin = np.ones(len(t))*Tjins
Tjin[60:100] = 230
```

```
# state vectors 
T = np.ones(len(t))*Ts
Tj = np.ones(len(t))*Tjs
Tsp = np.ones(len(t))*Ts
Tjsp = np.ones(len(t))*Tjs
```

```
# initial conditions 
s = [Ts, Tjs, Ts, Tjs]

# for cycle to change the inputs 
for i in range(len(t)-1):

    # update time span 
    ts = [t[i], t[i+1]]

    # update inputs 
    u[0] = Fj[i+1]
    u[1] = F[i+1]
    u[2] = Ti[i+1]
    u[3] = Tjin[i+1]

    # solve odes 
    out = odeint(heater, s, t, args=(u,))

    # update states 
    T[i+1] = out[-1][0]
    Tj[i+1] = out[-1][1]
    Tsp[i+1] = out[-1][2]
    Tjsp[i+1] = out[-1][3]

    # update initial conditions 
    s = out[-1]
```

```
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(t, T, t, Tj)
plt.xlabel("time (min)")
plt.legend(["T","Tj"])
plt.subplot(2,2,2)
plt.plot(t, Tsp, t, Tjsp)
plt.xlabel("time (min)")
plt.legend(["Tsp","Tjsp"])
plt.subplot(2,2,3)
plt.plot(t, F, t, Fj)
plt.xlabel("time (min)")
plt.legend(["F","Fj"])
plt.subplot(2,2,4)
plt.plot(t, Ti, t, Tjin)
plt.xlabel("time (min)")
plt.legend(["Ti","Tjin"])
```

* when Fj, Tj or Ti increase, the temperature in the tank and the jacket increase too

* when F increases, the temperature in the tank and the jacket decreasses

```
# nominal values
qs = 1000 # 1 std ft3/min
pis = 250 # psig
p1s = 50 # psig
p2s = 30 # psig
p3s = 5 # psig
his = 0.5
h1s = 0.5
h2s = 0.5
h3s = 0.5
Ts = 32 # °F pasarla a Rankine
```

```
# parameters
V1 = 1135 # ft3
V2 = 329 # ft3
R = 10.73 # psia ft3/(lbmol °R)
alphai = 10
alpha1 = 100
alpha2 = 80
```

```
def drum(s, t, u):

    # state variables 
    p1 = s[0]
    p2 = s[1]
    p1sp = s[2]
    p2sp = s[3]

    # inputs 
    h1 = u[0]
    h2 = u[1]
    pi = u[2]

    # physical model 
    # molar balance tank 1
    dp1dt = R*Ts/V1*(alphai*his*(pi-p1) - alpha1*h1*(p1-p2))
    # molar balance tank 2
    dp2dt = R*Ts/V2*(alpha1*h1*(p1-p2) - alpha2*h2*(p2-p3s))

    # state space model
    # molar balance tank 1 
    dp1spdt = -R*Ts/V1*(alphai*his + alpha1*h1s)*(p1sp - p1s) + R*Ts/V1*alpha1*h1s*(p2sp - p2s) -R*Ts/V1*alpha1*(p1s - p2s)*(h1 - h1s) + R*Ts/V1*alphai*his*(pi - pis)
    # molar balance tank 2
    dp2spdt = R*Ts/V2*alpha1*h1s*(p1sp - p1s) - R*Ts/V2*(alpha1*h1s + alpha2*h2s)*(p2sp - p2s) + R*Ts/V2*alpha1*(p1s - p2s)*(h1 - h1s) - R*Ts/V2*alpha2*(p2s - p3s)*(h2 - h2s)
    
    return [dp1dt, dp2dt, dp1spdt, dp2spdt]
```

```
# establish inputs 
u = [h1s, h2s+0.3, pis]

# establish initial conditions 
s = [p1s, p2s, p1s, p2s]
```

```
# integration time 
t = np.linspace(0,15,100)
```

```
# solve odes 
out = odeint(drum, s, t, args=(u,))
```

```
#@title plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(t, out[:,0],t, out[:,2])
plt.xlabel("time (min)", size = 15)
plt.ylabel("P1", size = 15)
plt.legend(["physical model","state space model"], fontsize = "x-large")
plt.subplot(1,2,2)
plt.plot(t, out[:,1],t, out[:,3])
plt.xlabel("time (min)", size = 15)
plt.ylabel("P2", size = 15)
plt.legend(["physical model","state space model"], fontsize = "x-large")
plt.subplots_adjust(bottom=0.1, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.35
)
```

```
# nominal values 
Fs = 100 # L/(mol hr)
#V1s = 400 # L
V1s = 2000
#V2s = 2000 # L
V2s = 400
c0s = 1 # mol/L
c1s = 1/3 # mol/L
c2s = 0.090052 # mol/L

# parameters 
k = 1.5 # L/(mol hr)
```

```
def two_reac(s,t,u):

    # state variables
    c1 = s[0]
    c2 = s[1]
    c1sp = s[2]
    c2sp = s[3]

    # inputs
    F = u[0]
    c0 = u[1]

    # component balance tank1 
    dc1dt = F/V1s*(c0 - c1) - k*c1**2
    # component balance tank2 
    dc2dt = F/V2s*(c1 - c2) - k*c2**2

    # component balance tank1 
    dc1spdt = -(Fs/V1s + 2*k*c1s)*(c1sp - c1s) + (c0s - c1s)/V1s*(F - Fs) + Fs/V1s*(c0 - c0s)
    # component balance tank2
    dc2spdt = Fs/V2s*(c1sp - c1s) -(Fs/V2s + 2*k*c2s)*(c2sp - c2s) +  (c1s - c2s)/V2s*(F - Fs)

    return [dc1dt, dc2dt, dc1spdt, dc2spdt]
```

```
# inputs 
u = [Fs, c0s]
#u = [Fs +10, c0s]
# initial conditions 
s = [c1s, c2s, c1s, c2s]
s1 = [0.3833, 0.09005, 0.3833, 0.09005]
s2 = [0.3333, 0.14005, 0.3333, 0.14005]
# integration time 
t = np.linspace(0,20,100)
print(eigenvec)
```
