Here you can see that when mu < 0 an unstable limit cycle between an stable node at 0 and an stable limit cycle close to 1. So, if we start close to zero we converge to zero, but if we start close to 1 the enter in an stable limit cycle. Let's check this :D

When mu = 0 or bigger than 0, there are only a stable node at zero and a stable limit cycle at 1. 

```
def model(s,t,mu):

    # state variable 
    r = s[0]
    theta = s[1]

    dr = mu*r + r**3 - r**5
    dtheta = -1

    return [dr, dtheta]
```

```
# intial condition 
s = [0.25, 0]
# integration time 
t = np.linspace(0,100,100)
```

```
mu = -0.1
# solve ODEs
out = odeint(model, s, t, args=(mu,))
```

```
#@title plot 
plt.figure(figsize= (12,10))
plt.subplot(2,2,1)
plt.plot(t,out[:,0])
plt.xlabel("time", size = 15)
plt.ylabel("r(t)", size = 15)
plt.subplot(2,2,2)
plt.plot(t,out[:,1])
plt.xlabel("time", size = 15)
plt.ylabel("theta(t)", size = 15)
# cartesian coordinates 
x = out[:,0]*np.cos(out[:,1])
y = out[:,0]*np.sin(out[:,1])
plt.subplot(2,2,3)
plt.plot(t,x,t,y)
plt.xlabel("time", size = 15)
plt.legend(["x(t)","y(t)"], fontsize = "x-large")
```

Now, let's start close to 1 

```
mu = 0.5
# solve ODEs
out = odeint(model, s, t, args=(mu,))
```

```
#@title plot 
plt.figure(figsize= (12,10))
plt.subplot(2,2,1)
plt.plot(t,out[:,0])
plt.xlabel("time", size = 15)
plt.ylabel("r(t)", size = 15)
plt.subplot(2,2,2)
plt.plot(t,out[:,1])
plt.xlabel("time", size = 15)
plt.ylabel("theta(t)", size = 15)
# cartesian coordinates 
x = out[:,0]*np.cos(out[:,1])
y = out[:,0]*np.sin(out[:,1])
plt.subplot(2,2,3)
plt.plot(t,x,t,y)
plt.xlabel("time", size = 15)
plt.legend(["x(t)","y(t)"], fontsize = "x-large")
```
